```python
%matplotlib inline
#Dependencies
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import statistics
#Load in csv
pyber_ride_df = pd.read_csv("Resources/PyBer_ride_data.csv")
pyber_ride_df
```

```python
pyber_ride_df.plot(x="Month", y="Avg. Fare ($USD)")
plt.show()
```

```python
# Set x-axis and tick locations.
x_axis = np.arange(len(pyber_ride_df))
tick_locations = [value for value in x_axis]
# Plot the data.
pyber_ride_df.plot(x="Month", y="Avg. Fare ($USD)")
plt.xticks(tick_locations, pyber_ride_df["Month"])
plt.show()
```

```python
# Chaining the bar() function to the plot() function to create a bar chart

pyber_ride_df.plot.bar(x="Month", y="Avg. Fare ($USD)")
plt.show()
```

```python
# Adding the kind parameter to the plot() function to create a bar chart:

pyber_ride_df.plot(x="Month", y="Avg. Fare ($USD)", kind='bar')
plt.show()
```

```python
# Skill Drill: Using the Data from the PyBer ride DataFrame, create a bar chart with the following annotations and characterics:
print(pyber_ride_df)

```

```python
#1. Add error bars:
# a) Determine standard deviation for "Avg. Fare in $USD" from data frame.
fare = pyber_ride_df["Avg. Fare ($USD)"]
std_fare = statistics.stdev(fare)
std_fare

      
```

```python
# b) Input into script with yerr=std_fare:

pyber_ride_df.plot(x="Month", y="Avg. Fare ($USD)", kind='bar', yerr=std_fare)
plt.show()
```

```python
#2. Add caps to the error bars
pyber_ride_df.plot(x="Month", y="Avg. Fare ($USD)", kind='bar', yerr=std_fare, capsize=5)
plt.show()
```

```python
#3. Change the color of the bars to sky blue.
pyber_ride_df.plot(x="Month", y="Avg. Fare ($USD)", kind='bar', yerr=std_fare, capsize=5, color="skyblue")
plt.show()
```

```python
#4. Rotate the labels on the x-axis to horizontal.
pyber_ride_df.plot(x="Month", y="Avg. Fare ($USD)", kind='bar', yerr=std_fare, capsize=5, color="skyblue")
xticks=pyber_ride_df["Month"]
plt.xticks(rotation=0)
plt.show()
```

```python
#5. Set the y-axis incement to every $5.

pyber_ride_df.plot(x="Month", y="Avg. Fare ($USD)", kind='bar', yerr=std_fare, capsize=5, color="skyblue")
xticks=pyber_ride_df["Month"]
plt.xticks(rotation=0)
plt.yticks(np.arange(0, 51, step=5.0))
plt.show()
```
