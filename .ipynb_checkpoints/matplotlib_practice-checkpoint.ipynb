```python
#Import dependencies
import matplotlib.pyplot as plt
```

```python
# Set the x-axis to a list of strings for each month.
x_axis = ["Jan", "Feb", "Mar", "April", "May", "June", "July", "Aug", "Sept", "Oct", "Nov", "Dec"]

# Set the y-axis to a list of floats as the total fare in US dollars accumulated for each month.
y_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]
```

```python
# Create the plot
plt.plot(x_axis, y_axis)
```

```python
# Using object-oritented interface method:
fig, ax = plt.subplots()
ax.plot(x_axis, y_axis)
```

```python
# Can also be written as: Create the plot with ax.plt()
fig, ax = plt.subplots()
ax.plot(x_axis, y_axis)
```

```python
# Create the plot with ax.plt()
fig = plt.figure()
ax = fig.add_subplot()

ax.plot(x_axis, y_axis)
```

```python
# Create the plot with ax.plt()
ax = plt.axes()
ax.plot(x_axis, y_axis)
```

```python
#To change figure-level attributes (such as axis labels, a title, or a legend) or save the figure as an image, use
fig = plt.figure()
```

```python
# Create the plot.
plt.plot(x_axis, y_axis)
plt.show()
```

```python
# Create the plot and add a label for the legend.
plt.plot(x_axis, y_axis, label='Boston')
# Create labels for the x and y axes.
plt.xlabel("Date")
plt.ylabel("Fare($)")
# Set the y limit between 0 and 45.
plt.ylim(0, 45)
# Create a title.
plt.title("PyBer Fare by Month")
# Add the legend.
plt.legend()
```

```python
# Create the plot.
plt.plot(x_axis, y_axis, marker="*", color="blue", linewidth=2, label='Boston')
# Create labels for the x and y axes.
plt.xlabel("Date")
plt.ylabel("Fare($)")
# Set the y limit between 0 and 45.
plt.ylim(0, 45)
# Create a title.
plt.title("PyBer Fare by Month")
# Add a grid.
plt.grid()
# Add the legend.
plt.legend()
```

```python
#Skill Drill using object-oriented approach:
# 1. Change line color to green
# 2. Change the marker to a diamond
# 3. Change the line width to 2 points
fig, ax = plt.subplots()
ax.plot(x_axis, y_axis,marker="d", color="green", linewidth=2, label='Boston')

# 4. Add a legend for the city of Boston.
ax.legend()

#5. Add a title and axes label.
ax.set_title("PyBer Fare By Month")
ax.set_xlabel("Date")
ax.set_ylabel("Fare($)")

# 6. Add a y-axis limit
ax.set_ylim(0,50)

#7. Add grid lines
ax.grid()
```

```python
# **Creating Bar Chart using MATLAB method**

# Set the x-axis to a list of strings for each month.
x_axis = ["Jan", "Feb", "Mar", "April", "May", "June", "July", "Aug", "Sept", "Oct", "Nov", "Dec"]

# Set the y-axis to a list of floats as the total fare in US dollars accumulated for each month.
y_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]
```

```python
# To Create a VERTICAL BAR CHART

#Create the plot
plt.bar(x_axis, y_axis)
```

```python
# To annotate the bar chart:

# Create the plot.
plt.bar(x_axis, y_axis, color="green", label='Boston')
# Create labels for the x and y axes.
plt.xlabel("Date")
plt.ylabel("Fare($)")
# Create a title.
plt.title("PyBer Fare by Month")
# Add the legend.
plt.legend()
```

```python
# To create a HORIZONTAL BAR CHART using MATLAB method.

# Create the plot
plt.barh(x_axis, y_axis)
```

```python
# To invert the axis:

plt.barh(y_axis, x_axis)
```

```python
# To invert y-axis to have months in ascending order:

#create the plot
plt.barh(x_axis, y_axis)
plt.gca().invert_yaxis()
```

```python
# Creating bar charts using Object Oriented Approach

# Set the x-axis to a list of strings for each month.
x_axis = ["Jan", "Feb", "Mar", "April", "May", "June", "July", "Aug", "Sept", "Oct", "Nov", "Dec"]

# Set the y-axis to a list of floats as the total fare in US dollars accumulated for each month.
y_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]
```

```python
# For data on opposite axes, need to switch arguments in the barh() function:

fig, ax = plt.subplots()
ax.barh(y_axis, x_axis)
```

```python
# Create a Bubble Chart Using the MATLAB method

plt.scatter(x_axis, y_axis, s=y_axis)
```

```python
y_axis_larger = []
for data in y_axis:
    y_axis_larger.append(data*3)
plt.scatter(x_axis, y_axis, s=y_axis_larger)
```

```python
# Using Python's list comprehension
# new_list = [expression for item in list if conditional]
plt.scatter(x_axis, y_axis, s = [i * 3 for i in y_axis])
```

```python
#Skill Drill: Using the object oriented approach, make the following changes to your bubble chart:

#1. Change the color of the markers to sky blue:
fig, ax=plt.subplots()
ax.scatter(x_axis,y_axis, s=y_axis, c="skyblue")

#2. Change the size of the markers to 5 times each data point.
ax.scatter(x_axis, y_axis, s = [i * 5 for i in y_axis], c="skyblue")

#3. Make the color 20% transparent.
ax.scatter(x_axis, y_axis, s = [i * 5 for i in y_axis], c="skyblue", alpha=0.2)

#4. Add a black edge around the circles.
ax.scatter(x_axis, y_axis, s = [i * 5 for i in y_axis], c="skyblue", alpha=0.2, edgecolor="black")

#5. Make the linewidth of the circles 2 points.
ax.scatter(x_axis, y_axis, s = [i * 5 for i in y_axis], c="skyblue", alpha=0.2, edgecolor="black", linewidth=2.0)

#6. Add a legend for the city of Boston.
ax.scatter(x_axis, y_axis, s = [i * 5 for i in y_axis], c="skyblue", alpha=0.2, edgecolor="black", linewidth=2.0, label="Boston")
ax.legend()

#7. Add a title
ax.set_title("PyBer Fare by Month")




#9. Add a limit for the x-axis data.

#10. Invert the y-axis so that January is at the top.

```

```python
#8. Swtich the axis so that the Fare($) data on the x-axis.
fig, ax=plt.subplots()
ax.scatter(y_axis, x_axis, s = [i * 5 for i in y_axis], c="skyblue", alpha=0.5, edgecolor="black", linewidth=2.0, label="Boston")
ax.legend()


#9. Add a limit for the x-axis data.
ax.set_xlim(0,50)

#10. Invert the y-axis so that January is at the top.
ax.set_title("PyBer Fare by Month")
ax.set_xlabel("Fare($)")
ax.set_ylabel("Date")
ax.invert_yaxis()
```
