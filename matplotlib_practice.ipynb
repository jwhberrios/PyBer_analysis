```python
#Import dependencies
import matplotlib.pyplot as plt
```

```python
# Set the x-axis to a list of strings for each month.
x_axis = ["Jan", "Feb", "Mar", "April", "May", "June", "July", "Aug", "Sept", "Oct", "Nov", "Dec"]

# Set the y-axis to a list of floats as the total fare in US dollars accumulated for each month.
y_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]
```

```python
# Create the plot
plt.plot(x_axis, y_axis)
```

```python
# Using object-oritented interface method:
fig, ax = plt.subplots()
ax.plot(x_axis, y_axis)
```

```python
# Can also be written as: Create the plot with ax.plt()
fig, ax = plt.subplots()
ax.plot(x_axis, y_axis)
```

```python
# Create the plot with ax.plt()
fig = plt.figure()
ax = fig.add_subplot()

ax.plot(x_axis, y_axis)
```

```python
# Create the plot with ax.plt()
ax = plt.axes()
ax.plot(x_axis, y_axis)
```

```python
#To change figure-level attributes (such as axis labels, a title, or a legend) or save the figure as an image, use
fig = plt.figure()
```

```python
# Create the plot.
plt.plot(x_axis, y_axis)
plt.show()
```

```python
# Create the plot and add a label for the legend.
plt.plot(x_axis, y_axis, label='Boston')
# Create labels for the x and y axes.
plt.xlabel("Date")
plt.ylabel("Fare($)")
# Set the y limit between 0 and 45.
plt.ylim(0, 45)
# Create a title.
plt.title("PyBer Fare by Month")
# Add the legend.
plt.legend()
```

```python
# Create the plot.
plt.plot(x_axis, y_axis, marker="*", color="blue", linewidth=2, label='Boston')
# Create labels for the x and y axes.
plt.xlabel("Date")
plt.ylabel("Fare($)")
# Set the y limit between 0 and 45.
plt.ylim(0, 45)
# Create a title.
plt.title("PyBer Fare by Month")
# Add a grid.
plt.grid()
# Add the legend.
plt.legend()
```

```python
#Skill Drill using object-oriented approach:
# 1. Change line color to green
# 2. Change the marker to a diamond
# 3. Change the line width to 2 points
fig, ax = plt.subplots()
ax.plot(x_axis, y_axis,marker="d", color="green", linewidth=2, label='Boston')

# 4. Add a legend for the city of Boston.
ax.legend()

#5. Add a title and axes label.
ax.set_title("PyBer Fare By Month")
ax.set_xlabel("Date")
ax.set_ylabel("Fare($)")

# 6. Add a y-axis limit
ax.set_ylim(0,50)

#7. Add grid lines
ax.grid()
```

```python
# **Creating Bar Chart using MATLAB method**

# Set the x-axis to a list of strings for each month.
x_axis = ["Jan", "Feb", "Mar", "April", "May", "June", "July", "Aug", "Sept", "Oct", "Nov", "Dec"]

# Set the y-axis to a list of floats as the total fare in US dollars accumulated for each month.
y_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]
```

```python
# To Create a VERTICAL BAR CHART

#Create the plot
plt.bar(x_axis, y_axis)
```

```python
# To annotate the bar chart:

# Create the plot.
plt.bar(x_axis, y_axis, color="green", label='Boston')
# Create labels for the x and y axes.
plt.xlabel("Date")
plt.ylabel("Fare($)")
# Create a title.
plt.title("PyBer Fare by Month")
# Add the legend.
plt.legend()
```

```python
# To create a HORIZONTAL BAR CHART

# Create the plot
plt.barh(x_axis, y_axis)
```

```python
# To invert the axis:

plt.barh(y_axis, x_axis)
```

```python
# To invert y-axis to have months in ascending order:

#create the plot
plt.barh(x_axis, y_axis)
plt.gca().invert_yaxis()
```

```python
#Skill Drill: Using Matplotlib MATLAB approach, make the following changes:

#1. Change the car colors to magenta:
plt.barh(x_axis, y_axis, color="magenta", label="Boston")
plt.gca().invert_yaxis()

#2. Add a legend for the city of Boston.
plt.legend()

#3. Add a title and axes labels.
plt.title("PyBer Fare By Month")
plt.xlabel("Fare($)")
plt.ylabel("Date")


```

```python
# Creating bar charts using Object Oriented Approach

# Set the x-axis to a list of strings for each month.
x_axis = ["Jan", "Feb", "Mar", "April", "May", "June", "July", "Aug", "Sept", "Oct", "Nov", "Dec"]

# Set the y-axis to a list of floats as the total fare in US dollars accumulated for each month.
y_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]
```

```python
# Create the plot with ax.plt()

fig, ax = plt.subplots()
ax.bar(x_axis, y_axis)

#To change bar graph color to black with a legend for the city of New York:
ax.bar(x_axis, y_axis, color="black", label="New York")
```

```python
#Create a Horizontal Bar Chart
fig, ax = plt.subplots()
ax.barh(x_axis, y_axis)
```

```python
# For data on opposite axes, need to switch arguments in the barh() function:

fig, ax = plt.subplots()
ax.barh(y_axis, x_axis)
```
